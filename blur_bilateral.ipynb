```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('gaussian_noise.jpg')
```

```python
#가우시안 필터 적용
blur1 = cv2.GaussianBlur(img, (5,5), 0)
```

```python
#바이레터럴 필터 적용
blur2 = cv2.bilateralFilter(img, 5, 75, 75)
```

```python
#결과 출력
merged = np.hstack((img, blur1, blur2))
cv2_imshow(merged)
```

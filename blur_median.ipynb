```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('salt_pepper_noise.jpg')
```

```python
#미디언 블러 적용
blur = cv2.medianBlur(img, 5)
```

```python
#결과 출력
merged = np.hstack((img, blur))
cv2_imshow(merged)
```
